# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
def Relu(x):
    return np.maximum(0, x)

def dRelu(x):
    return np.where(x>0, 1, 0)
#     return lambda x: 1 if x > 0 else 0
```

```python
x=np.arange(-1.0,1.0,0.01)
# x = plt.linspace(-5,5,100)
print('ReLu\n' + str(Relu(x)))
print('dReLu\n' + str(dRelu(x)))
```
